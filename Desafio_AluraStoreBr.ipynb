### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
#Análise sobre o faturamento das lojas.
faturamento_loja1 = loja ['Preço'].sum()
faturamento_loja2 = loja2 ['Preço'].sum()
faturamento_loja3 = loja3 ['Preço'].sum()
faturamento_loja4 = loja4 ['Preço'].sum()

print(f"Loja 1: R$ {faturamento_loja1: .2f}")
print(f"Loja 2: R$ {faturamento_loja2: .2f}")
print(f"Loja 3: R$ {faturamento_loja3: .2f}")
print(f"loja 4: R$ {faturamento_loja4: .2f}")
```

```python
# Mostrando gráfico sobre faturamento das lojas.
import matplotlib.pyplot as plt
lojas = ["loja 1", "loja 2", "loja 3", "loja 4"]
faturamento = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

plt.figure(figsize = (10, 5))
barras = plt.bar (lojas, faturamento, color = 'green')
plt.title("Faturamento das Lojas")
plt.ylabel("Faturamento (R$)")
plt.xlabel("Lojas")
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
for barra, valor in zip (barras, faturamento):
    plt.text(barra.get_x() + barra.get_width() / 2, barra.get_height() + 20000, f'R$ {valor:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'),
             ha='center', va='bottom', fontsize = 9, fontweight='bold')
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
from collections import Counter
avaliacao = Counter(loja['Avaliação da compra'].tolist())
avaliacao_2 = Counter(loja2['Avaliação da compra'].tolist())
avaliacao_3 = Counter(loja3['Avaliação da compra'].tolist())
avaliacao_4 = Counter(loja4['Avaliação da compra'].tolist())

def media_avaliacao(loja):
    avaliacao = loja['Avaliação da compra'].tolist()
    return sum(avaliacao) / len(avaliacao) if avaliacao else 0
print(f'Média da Avaliação Loja 1: {media_avaliacao(loja):.2f}')
print(f'Média da Avaliacao Loja 2: {media_avaliacao(loja2):.2f}')
print(f'Média da Avaliação Loja 3: {media_avaliacao(loja3):.2f}')
print(f'Média da Avaliação Loja 4: {media_avaliacao(loja4):.2f}')
```

```python
# Gráfico Sobre Avaliação das lojas
import matplotlib.pyplot as plt
lojas = ["loja 1", "loja 2", "loja 3", "loja 4"]
media = [media_avaliacao(loja), media_avaliacao(loja2), media_avaliacao(loja3), media_avaliacao(loja4)]

plt.figure(figsize = (10, 5))
plt.plot(lojas, media, marker='o', color = 'red', linewidth = 2, linestyle = '--')
plt.title('Média de Avaliação das Lojas')
plt.ylabel('Média de Avaliação')
plt.xlabel('Lojas')
plt.ylim(0,5)
plt.grid(True, linestyle = '--', alpha = 0.7)

for i, media in enumerate(media):
  plt.text(lojas[i], media + 0.1, f'{media:.2f}', ha ='center', fontsize = 9, fontweight = 'bold')

plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
from ast import Lambda
from collections import Counter
produtos = Counter(loja['Produto'])
produtos2 = Counter(loja2['Produto'])
produtos3 = Counter(loja3['Produto'])
produtos4 = Counter(loja4['Produto'])

# Função para produtos mais e menos vendidos
def mais_menos_vendidos(nome_loja, contador):
    if not contador:
        print(f"{nome_loja}: Sem dados de vendas.")
        return

    mais_vendido = contador.most_common(1)[0]
    menos_vendido = min(contador.items(), key=lambda x: x[1])

    print(f"\n {nome_loja}")
    print(f"  Produto mais vendido: {mais_vendido[0]} ({mais_vendido[1]} vendas)")
    print(f"  Produto menos vendido: {menos_vendido[0]} ({menos_vendido[1]} vendas)")

mais_menos_vendidos("Loja 1", produtos)
mais_menos_vendidos("Loja 2", produtos2)
mais_menos_vendidos("Loja 3", produtos3)
mais_menos_vendidos("Loja 4", produtos4)
```
